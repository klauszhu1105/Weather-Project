```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from xgboost import XGBRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
# Load the dataset
file_path = "/Users/zhuyun/Downloads/archive/GlobalWeatherRepository.csv"
df = pd.read_csv(file_path)
```

```python
# Identify numerical columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Handle missing values
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
# Detect and handle outliers using IQR method
def remove_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    return df

df = remove_outliers_iqr(df, numeric_cols)
```

```python
# Normalize data using Min-Max Scaling
scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
# Ensure 'last_updated' is in datetime format
df['last_updated'] = pd.to_datetime(df['last_updated'])

# Set it as index
df.set_index('last_updated', inplace=True)

# Select only numeric columns for resampling
numeric_cols = df.select_dtypes(include=['number']).columns

# Interpolate missing values before resampling
df[numeric_cols] = df[numeric_cols].interpolate(method='time')

# Resample to daily frequency and take the mean
df_resampled = df[numeric_cols].resample('D').mean()

# Fill any remaining NaNs using forward and backward fill
df_resampled = df_resampled.fillna(method='ffill').fillna(method='bfill')


# Check missing values again
print(df[numeric_cols].isna().sum())

# Select temperature column for forecasting
target_col = 'temperature_celsius'
```

```python
# Select temperature column for forecasting
target_col = 'temperature_celsius'
df
```

```python
# Split data into training and testing sets (80% train, 20% test)
train_size = int(len(df_resampled) * 0.8)
train, test = df_resampled[:train_size], df[train_size:]
```

```python
# Function to evaluate model performance
def evaluate_forecast(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    return {'Model': model_name, 'MAE': mae, 'RMSE': rmse}
# Store results
results = []
```

```python
# ARIMA Model
arima_model = ARIMA(train[target_col], order=(5, 1, 0))
arima_model_fit = arima_model.fit()
arima_pred = arima_model_fit.forecast(steps=len(test))
results.append(evaluate_forecast(test[target_col], arima_pred, "ARIMA"))
```

```python
# XGBoost Regressor
x_train, x_test, y_train, y_test = train_test_split(
    np.arange(len(train)).reshape(-1, 1), train[target_col], test_size=0.2, shuffle=False
)

xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1)
xgb_model.fit(x_train, y_train)

xgb_pred = xgb_model.predict(np.arange(len(train), len(train) + len(test)).reshape(-1, 1))
results.append(evaluate_forecast(test[target_col], xgb_pred, "XGBoost"))
```

```python
# LSTM Model
look_back = 5

def create_sequences(data, look_back):
    X, Y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i : i + look_back])
        Y.append(data[i + look_back])
    return np.array(X), np.array(Y)

X_train, y_train = create_sequences(train[target_col].values, look_back)
X_test, y_test = create_sequences(test[target_col].values, look_back)

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

lstm_model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(look_back, 1)),
    LSTM(50),
    Dense(1)
])
lstm_model.compile(optimizer=Adam(learning_rate=0.01), loss='mse')

lstm_model.fit(X_train, y_train, epochs=20, batch_size=8, verbose=0)
lstm_pred = lstm_model.predict(X_test).flatten()
results.append(evaluate_forecast(y_test, lstm_pred, "LSTM"))
```

```python
# Convert results to DataFrame and display
results_df = pd.DataFrame(results)
print(results_df)
```
